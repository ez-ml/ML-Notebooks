    <b>ASIDE:</b> 
    What is wrong with modeling the data as follows? This is a normal curve with mean 33.66 and stdev=97.02.

**Aside Answer:** There are many problems with this. First the sample data are not normally distributed as we saw above in the green histogram, and having orders with negative values is nonsensical. So how is this not a problem for doing hypothesis testing, if our data are not normally distributed?

The answer lies in what we are testing. We are not testing whether individual order values are normally distributed. We want to know about the average (mean) order value. It is the **population mean** that we want to estimate, and we can estimate it and make decisions about it based on the mean of the sample data - the **sample mean**.

This is where it is critical to understand the difference between the standard deviation of the sample data and the standard deviation of our estimate of the sample mean, which is known as the standard error of the sample mean. Consider the following curve:

<img src="images/graph2.png" style="width: 450px; height: 250px;" />

This is a normal curve based on the sample mean, with standard deviation equal to the standard error of the mean. (Standard error = standard deviation divided by the square root of the number of samples.) **This is the curve that we can use to talk about confidence intervals.**

Why is it OK to assume the data here are normal, when we know that the Order values were NOT normally distributed? The answer is that the Central Limit Theorem tells us that statistics such as the sample mean are normally distributed (for large samples) even when the underlying data are not normally distributed. (I hope you stop for a moment to consider how powerful this conclusion is. It is the basis of most statistical inference in our universe.)
<br><br>
Now, we can take this curve and shift it to some H0 that we want to test, we can then make conclusions about how the population mean.
<br><br>
For example, if we decided the the break-even order amount was $10, then we would draw the curve like this:


<img src="images/graph3.png" style="width: 450px; height: 250px;" />

Then we would state our “business-as-usual” null hypothesis as: 

H<sub>0</sub>: The population mean order is less than or equal to $10.

Here we are making our test explicitly a one-tailed test. That is, we are looking for some strong reason to buy the list - the reason being that we are confident that it would make money for the company. We don’t care about the left tail of the curve as that’s a money-loser. So we focus on that 5\% (assuming 95\% confidence) area under the curve to the right tail only.

Then we would look at the dotted red line, and ask “How unusual would it be to get data like the dotted red line from a data sample if, in fact, the dotted blue line is the true population mean?”

Think about this. We don’t want to buy the list if there is some reasonable chance that we will lose money on it. If that’s the case, then we want business- as-usual, and our null hypothesis defends us from that.

To make this determination, we would compute a t-statistic for the red line at the desired confidence level, say 95\%, and then make a decision. Our decision can have only two forms:

1. Our data, the red line, is so unusual (statistically unlikely), that at the 95\% level, we reject the idea (null hypothesis) that the average order is in fact $10 or less. Reject the null, no business as usual, buy the list.
<br><br>

2. Our data, the red line, is not that unusual, and in fact, it *does* fall in the 95\% probability zone, and therefore, we would expect that more than 5\% of the time, we could get data like this if the true population mean is \\$10. So we conclude that there’s too big a chance that buying the list will NOT be profitable for us. Fail to reject the null. Do not buy the list, and proceed business-as-usual.

There is no such thing as accepting the null hypothesis. If this experiment were repeated 1000 times and we looked at the most extreme set of sample data from those tests, we might find that the data in the sample data were highly unusual and fell outside of even a 99% confidence interval, but that were still a product of the random sampling. It is for this reason that we always mention the confidence level when we state a conclusion of a hypothesis test, and that we always either reject the null, or fail to reject the null.
The p-value is the calculation of how bizarre it would be to get the data we got if the null hypothesis were true. For example is p=0.003, then only in 0.3% of samples would we get such data if the null hypothesis were true.
